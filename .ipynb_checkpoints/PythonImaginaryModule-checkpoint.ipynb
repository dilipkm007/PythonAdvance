```python
from PIL import Image, ImageFilter, ImageFont, ImageDraw
from matplotlib.pyplot import imshow
```

```python
%ls -l
```

```python
file_name = "SierinkshiTriangle.png"
im = Image.open(file_name)
imshow(im)
```

```python
print(im.size)
print(im.format_description)
print(im.mode)
```

```python
crop = (500, 500, 800, 800)
croped_image = im.crop(crop)
imshow(croped_image)
```

```python
resized_image = im.resize((400, 400))
imshow(resized_image)
```

```python
gray_scale = im.convert('LA')
imshow(gray_scale)
```

```python
effect = im.filter(ImageFilter.GaussianBlur(radius=15))
imshow(effect)
```

```python
draw = ImageDraw.Draw(im)
font = ImageFont.truetype('halkduster.ttf', 200)
draw.text((100, 100), "Giles' Photo!", font = font)
imShow(im)
```
