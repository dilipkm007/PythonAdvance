## Random walk

```python
def rand_walk(step_num):
    walk = []
    step_choice = choice([-1, 1])
    walk.append(step_choice)
    for i in range(1, step_num):
        choice2 = choice([-1, 1])
        next_Step = walk[i-1] + choice2
        walk.append(next_Step)
    return walk
        
        
```

```python
def plot_walk(walk):
    x = list(range(1, len(walk)+1))
    plt.plot(x, walk)
    plt.xlabel('Number of Steps')
    plt.ylabel('Distance from origin')
    plt.title('Our random walk')
    
```

```python
plot_walk(rand_walk(1000))
```

```python
def plot_walks(num, walk_steps):
    lab_list = list(range(1, num+1))
    for i in range(0, num):
        x = list(range(1, walk_steps + 1))
        plt.plot(x, rand_walk(walk_steps), label = 'Plot number' + ' ' + str(lab_list[i]))
        plt.xlabel('Number of Steps')
        plt.ylabel('Distance from origin')
        plt.title('Our random walk of ' + str(num) + " days.")
        plt.legend(loc='lower left')
        plt.savefig("Randomwalk.png")
    plt.show()
```

```python
plot_walks(7, 100)
```

## Sierinkshi Triangle 
First Trnsport: x<sub>n+1</sub> = 0.5x<sub>n</sub><br>y<sub>n+1</sub> = 0.5y<sub>n</sub>

Second Trnsport: x<sub>n+1</sub> = 0.5x<sub>n</sub> + 0.5<br>y<sub>n+1</sub> = 0.5y<sub>n</sub> + 0.5


Third Trnsport: x<sub>n+1</sub> = 0.5x<sub>n</sub> + 1<br>y<sub>n+1</sub> = 0.5y<sub>n</sub>


```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1, y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1, y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1.0
    y1 = 0.5 * y
    return x1, y1
```

```python
transformations = [trans_1, trans_2, trans_3]
```

```python
a1 = [0]
b1 = [0]
a, b = 0, 0
```

```python

for i in range(1000000):
    trans = choice(transformations)
    a, b = trans((a, b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure', figsize = (16, 16))
```

```python
plt.plot(a1, b1, 'o')
plt.savefig("SierinkshiTriangle .png")
```
